```python
import pandas as pd
pd.options.display.max_columns = None # show all columns
import numpy as np
import missingno as msno 
# Jeżeli by się nie importowało spróbujcie w terminalu anacondy wpisać conda install --channel "conda-forge" missingno
import seaborn as sns # wizualizacja danych
import matplotlib.pyplot as plt # wykresy
import collections
import random

data1 = pd.read_csv('wine_white_1.csv', sep=';')
data2 = pd.read_csv('wine_white_2.csv', sep=';')
data3 = pd.read_csv('wine_white_3.csv', sep=';')
dataNew = pd.concat([data1, data2, data3], axis=1, sort=False)
```

```python
dataNew.head()
```

```python
correlation = dataNew.corr()
print(correlation['quality'].sort_values(ascending = False))
```

```python
# Mapa cieplna korelacji atrybutów
f , ax = plt.subplots(figsize = (14,12))
# plt.subplots(figsize = (14,12))
plt.title('Korelacja atrybutów numerycznych',size=15)
sns.heatmap(correlation,square = True,  vmax=0.8)
```

```python
# Zbliżenie mapy cieplnej najbardziej skorelowanych atrybutów
zoomedCorrelation = correlation.loc[['residual sugar', 'density', 'free sulfur dioxide', 'total sulfur dioxide'], ['residual sugar', 'density', 'free sulfur dioxide', 'total sulfur dioxide']]
f , ax = plt.subplots(figsize = (7,6))
plt.title('Korelacja atrybutów numerycznych',size=10)
sns.heatmap(zoomedCorrelation, square = True, linewidths=0.01, vmax=1, annot=True,cmap='viridis',
            linecolor="black", annot_kws = {'size':12})
```

```python
columns = list(dataNew.columns)

sns.set()
sns.pairplot(dataNew[columns],height=2 ,kind ='scatter',diag_kind='kde')
plt.show()
```

```python
plt.plot(dataNew['density'], dataNew['residual sugar'], 'ro')
plt.title('residual sugar a density',size=10)
plt.show()
```

```python
# Visualising missing values of numeric features for sample of 1000
msno.matrix(dataNew.sample(1000))
```

```python
# Sprawdzamy ilość brakujących obserwacji
dataNew['residual sugar'].isnull().sum()
```

```python
# Dzielimy 'YearBuilt' na 10 przedziałów
dataNew['density_cut'] = pd.qcut(dataNew.density,10)
dataNew
```

```python
residual_sugar_grouped = dataNew.groupby(['density_cut'])['residual sugar']

dataNew['residual sugar']= residual_sugar_grouped.transform(lambda x : x.fillna(x.median()))

dataNew['residual sugar'] = dataNew['residual sugar'].astype(int)

dataNew.drop('density_cut',axis=1,inplace=True)
```

```python
# Sprawdzamy czy udało nam się zastąpić brakujące obserwacje.
dataNew['residual sugar'].isnull().sum()
```

```python
dataNew.to_csv('TUTAJ PODAJESZ ŚCIEŻKĘ')
```

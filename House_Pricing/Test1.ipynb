```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
#Creates just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```
